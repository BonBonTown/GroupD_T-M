```python
import pandas as pd

# 读取Excel文件
data = pd.read_excel('豆瓣电影成果.xlsx')

# 显示数据的前几行
data.head()
```

```python
split_data = data['演职人员'].str.split('\n').tolist()

# 创建新的DataFrame，将分割后的字符串放入新的单元格
new_data = pd.DataFrame(split_data)

new_data
```

```python
crew_data= pd.read_csv('香港金像奖获奖演员导演信息.csv', sep=';')

crew_data.drop_duplicates(subset='Name', inplace=True)

crew_data.head()
```

```python
from itertools import combinations

result_data = pd.DataFrame(columns=['name_value_1', 'name_value_2', 'weight'])

# Generate combinations of every pair of names
all_name_combinations = list(combinations(crew_data['Name'], 2))
unique_combinations = []

for name1, name2 in all_name_combinations:
    # Sort names to ensure consistent ordering
    sorted_names = sorted([name1, name2])
    if sorted_names not in unique_combinations:
        unique_combinations.append(sorted_names)
        
# Iterate through each pair of names
for name1, name2 in all_name_combinations:
    weight = 0

    # Iterate through each row in data
    for _, row in data.iterrows():
        row_names = row['演职人员'].split('\n')
        
        # Check if both names appear in the same row (partial match)
        if any(name1 in row_name for row_name in row_names) and any(name2 in row_name for row_name in row_names):
            weight += 1
    
    # Print the count of each pair of names
    print(f"Names: {name1}, {name2} - Weight: {weight}")
    
    # If the pair of names appeared together, add them to the result DataFrame
    if weight > 0:
        result_data = pd.concat([
            result_data,
            pd.DataFrame({
                'name_value_1': [name1],
                'name_value_2': [name2],
                'weight': [weight]
            })
        ], ignore_index=True)

print(result_data)
```

```python
result_data.to_csv('result_data.csv', index=False)
```
